**9.** Let $X_1, \dots, X_n \sim N(0,1)$. Plot $\overline{X}_n$ versus $n$ for $n = 1, \dots, 10,000$. Repeat for $X_1, \dots, X_n \sim \text{Cauchy}$. 

```python
n = 10_000
```

```python
x = rng.normal(size=n)
xbar = np.cumsum(x) / np.arange(1,n+1)
plt.plot(np.arange(1,n+1), xbar, 'o')
plt.show()
```

```python
from scipy.stats import cauchy
y = cauchy.rvs(size=n, random_state=rng)
ybar = np.cumsum(y) / np.arange(1,n+1)
plt.plot(np.arange(1,n+1), ybar, 'o')
plt.show()
```

**11b.** Let $Y_1, Y_2, \dots$ be independent random variables such that $P(Y_i = 1) = P(Y_i = -1) = 1/2$. Let $X_n = \sum_{i=1}^n Y_i$. Simulate $X_n$ and plot $X_n$ versus $n$ for $n = 1, \dots, 10,000$. Repeat the whole simulation several times. 

```python
for seed in range(0,4):
    rng = np.random.default_rng(seed)
    z = rng.choice([-1,1], size=n)
    w = np.cumsum(z)
    plt.plot(np.arange(1,n+1), w)
    plt.show()
```

**19.** Let $X_1, \dots, X_n \sim U(0,1)$. Simulate the distribution of the sample mean $\overline{X}_n = \sum_{i=1}^n X_i$ for $n=1,5,25,1000$. Compare the simulated values of $E \overline{X}_n$ and $V \overline{X}_n$ with the theoretical ones.

```python
for n in [1,5,25,1000]:
    u_bar = [np.mean(rng.uniform(0, 1, n)) for _ in range(10000)]
    plt.hist(u_bar, bins=33, density=True)
    plt.xlim(0,1)
    plt.show()
    print('n = ' + str(n))
    print('Mean difference: ' + str(np.mean(u_bar) - 0.5))
    print('Variance difference: ' + str(np.var(u_bar) - 1 / (12*n)))

# Alternatively,
# for n in [1, 5, 25, 1000]:
    # u = np.random.uniform(0, 1, (10000, n))
    # u_bar = u.mean(axis=1)
```
