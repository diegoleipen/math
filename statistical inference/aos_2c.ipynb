```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

**13b.** Generate a random vector $X = (x_1, \dots, x_{10,000})$ consisting of 10,000 random standard normals. Let $y = (e^{x_1}, \dots, e^{x_{10,000}})$. Draw a histogram of $y$ and compare it to the PDF of $Y = e^X$, where $X \sim N(0,1)$.

```python
x = np.random.normal(0, 1, size=10000)
y = np.exp(x)

bins = np.linspace(0, 10, 33)
plt.hist(y, bins=bins, density=True, edgecolor="black")

w = np.linspace(0.01, 10, 1000)
pdf = (1/w) * norm.pdf(np.log(w))

plt.plot(w, pdf)
plt.xlim(0, 10)
plt.show()
```

**15.** Write a program that takes $\text{Uniform}(0,1)$ random variables and generates random variables from an $\text{Exponential}(\beta)$ distribution.

```python
u = np.random.uniform(0,1,10000)

beta = 0.3
y = - beta * np.log(1-u)

bins = np.linspace(0, 2, 33)
plt.hist(y, bins=bins, density=True, edgecolor="black")

x = np.linspace(0.01, 2, 1000)
pdf = 1/beta * np.exp(-x/beta)

plt.plot(x, pdf)
plt.xlim(0, 2)
plt.show()
```
